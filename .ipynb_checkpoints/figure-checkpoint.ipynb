```python
def get_energy_per_mode(k_vector, rad_energy):
    k_vector = np.array(k_vector)
    rad_energy = np.array(rad_energy)
    
    k_vector = k_vector[:,0,:]
    
    k_val = np.sqrt(np.einsum("ki,ki->k",k_vector,k_vector))
    k_val = np.round(k_val, decimals = 8)
    
    unique_kval = np.sort(np.array(list(set(k_val))))
    
    final_rad_energy = rad_energy[-1,:]
    
    k_rad_energy = []
    
    for k in unique_kval:
        k_rad_energy.append(np.sum(final_rad_energy[k_val == k]))
    
    k_rad_energy = np.array(k_rad_energy)

    return unique_kval, k_rad_energy 

```
